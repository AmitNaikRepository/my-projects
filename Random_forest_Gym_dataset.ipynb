```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split 
```

```python
from subprocess import check_output
from datetime import time
#print(check_output(["ls","C:\Users\amit\Desktop\practice\gym.csv"]).decode("utf8"))
```

```python
def time_to_seconds(time):
    return time.hour*3600+ time.minute *60 + time.second
```

```python
df = pd.read_csv(r"C:\Users\amit\Desktop\practice\gym.csv")
# Drop columns
df = df.drop("date", axis=1)
# center timestamp
noon = time_to_seconds(time(12, 0, 0))
df.timestamp = df.timestamp.apply(lambda t: abs(noon - t))
columns = ["day_of_week", "month", "hour"]
df = pd.get_dummies(df, columns=columns)
df.head(10)
```

```python
#extract trainning and testing data value we used will get data into data frame [X,y]
data=df.values
X=data[:,1:]
y=data[:,0]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#now we scale the feature between the -1 to 1
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#create a model npw
model =RandomForestRegressor()
```

```python
#now we take n_estimator it usually tell how many decision tree that we will used 
estimators=np.arange(10,200,10)
score=[]
for n in estimators:
    model.set_params(n_estimators=n)
    model.fit(X_train,y_train)
    score.append(model.score(X_test,y_test))
plt.label("effect of n estimators ")
plt.xlabel="n_estimators"
plt.ylabel="scores"
plt.plot(estimators ,score )
```
