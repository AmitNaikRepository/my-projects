```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
import os

DATADIR=(r"C:\Users\amit\Desktop\datasets\PetImages")
CATEGORIES=['dog','cat']

for category in CATEGORIES:
    path=os.path.join(DATADIR,category) 
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array,cmap='gray')
        plt.show()
        break
    break
```

```python
print(img_array.shape)
#now the next step is to normalize how the dataset
```

```python
IMG_SIZE=80
new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_array,cmap='gray')
plt.show()
```

```python
#now lets do the training data 
training_data=[]
def create_training_data():
    for category in CATEGORIES:
        path=os.path.join(DATADIR,category) 
        class_num=CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
create_training_data()
```

```python
print(len(training_data))
```

```python
#now lets shuffle the data
import random
random.shuffle(training_data) 
#now lets check if the training data is shuffle 
for sample in training_data[:10]:
    print(sample[1])
    #we use print 1 beacuse we need to see what label the sample hold 
```

```python
X=np.array([])
y=np.array([])
```

```python
for features,label in training_data:
    X=np.append(X,features)
    y=np.append(y,label)
#so now we can not serv e the list of for training datasets we must transfer to numpy array so ;lets do this
X=np.array(X).reshape(-1, IMG_SIZE,IMG_SIZE,1)
```

```python
import pickel  
pickle_out=open("x.pickel","wb")
pickle.dump(X,pickle_out)
pickle_out.close()

pickle_out=open("y.pickel","wb")
pickle.dump(y,pickle_out)
pickle_out.close()
```
