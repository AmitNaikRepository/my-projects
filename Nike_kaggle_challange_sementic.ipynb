```python
random.seed(214)
text=tweets.tweet_full_text[random.sample(range(1,80),10)]
text
```

```python
#now lets take the lingustic annotaion waht does it measn is simpple the noun adjective and others are categorise n the text file 
text_combined=str(text)
doc=nlp(text_combined)
```

```python
#so now we do tokenize the anotation its called chunking in nlp 
for token in doc:
    print(token)
```

```python
#now lets prints with theire print tags 
for token in doc:
    print(token.text,token.pos_)
```

```python
#now we can see that each word got its pos tags defind
#now we will print only noun
nouns=list(doc.noun_chunks)
nouns 
```

```python
#now lets transcript the sentence okay 
list(doc.sents)
```

```python
#now we will the do (NER)which means it will excract the named entities 
for ent in doc.ents:
    print(ent.text,ent.label_)
```

```python
#now lesta take the worth of the text so we will take text label spacy can do thiss letws try 
spacy.displacy.render(doc,style='ent',jupyter=True)
```

```python
#now we will do lmmatization which is used to display the word
for token in doc:
    print(token.text,token.lemma_)
```

```python
#now lets take dependacny parser visualization 
spacy.displacy.render(doc, style='dep',jupyter=True)
```
